```python
# countries mentions in scientific papers 
scientific_mentions = pd.read_excel(r'C:\Users\charb\Downloads\scimagojr.xlsx')

print(energy[energy.Country.str.contains("[0..9]")].Country)


scientific_mentions.head()
```

```python
# Three way step merging, and then select the top 15 highest by Rank column
gdp = gdp[['Country Name', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']].set_index('Country Name')
nrg = energy[['Country','Energy Supply', 'Energy Supply per Capita', '% Renewable']].set_index('Country')
sci = scientific_mentions[['Rank','Country', 'Documents', 'Citable documents', 'Citations', 
              'Self-citations', 'Citations per document', 'H index']].set_index('Country')
scinrg=pd.merge(sci,nrg,how='inner', left_index = True, right_index = True)
merged = pd.merge(scinrg,gdp, how='inner', left_index = True, right_index = True)
res = merged.sort_values('Rank',ascending=True).head(15)

res
```

```python
scinrgo=pd.merge(sci,nrg,how='outer', left_index = True, right_index = True)
mergedo = pd.merge(scinrgo,gdp, how='outer', left_index = True, right_index = True)

mergedo.shape[0] - merged.shape[0]
```
